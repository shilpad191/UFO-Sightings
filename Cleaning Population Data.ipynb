# Number of Sightings vs Increase in Population Data

We begin by reading the total number of sightings per year from the cleaned data sightings_per_year.csv. This cleaned datset contains two columns: 
- 'Year', which is also the index 
- 'Posted' which is the total sightings posted count

```python
yearlypop = pd.read_csv(r'clean_data/Yearly_sightings', index_col='Year')
```

```python
yearlypop.head()
```

We then read the total population by year dataset. This dataset contains two columns:
- Year, which is also the index
- and 'Population'

```python
Pop = pd.read_csv(r'Raw_Data/Population-DS-Year.csv', index_col='Year')
```

```python
Pop.head()
```

We now join the above two datasets by using join in order to create a new dataframe called Population_Sightings. 

```python
Population_Sightings = Pop.join(yearlypop)
Population_Sightings = Population_Sightings.dropna()
Population_Sightings['Population'] = Population_Sightings['Population'] / 1000000
Population_Sightings.head()
```

The above dataframe is stored into a csv file 'Sightings_by_population_yearly' for later user to plot a graph to find the trend between increase in the number of sightings with the increase in population.

```python
Population_Sightings.to_csv(r'Clean_data/Sightings_by_population_yearly', header=True)
```

# Sightings Per Person Data

We begin by reading the total number of sightings in a state from the cleaned data Sightings_by_state.csv. This cleaned datset contains two columns: 
- 'STATE_ABBR', which is also the index 
- 'Posted' which is the total sightings reported in the state

```python
sightingsbystate = pd.read_csv(r'clean_data/Sightings_by_state', names = ['STATE_ABBR', 'Posted'], index_col ='STATE_ABBR')
sightingsbystate.head()
```

We then read the total population by state from the cleaned data Population-Dataset-states_abbr.csv. This cleaned datset contains two columns: 
- 'State', which is also the index 
- 'Population', which is the population of the state

```python
populationbystate = pd.read_csv(r'Raw_Data/Population-Dataset-states_abbr.csv', index_col ='State')
populationbystate.head()
```

We now join the above two datasets by using join in order to create a new dataframe called State_Population_Sightings. 

```python
State_Population_Sightings = populationbystate.join(sightingsbystate, lsuffix='l', rsuffix='r' )
State_Population_Sightings.head()
```

We then create a new column called SightingPerPerson and calculate its value for each state as the total sightings reported in the state divided by the total population in the state multiplied by 10000. This gives us the total number of sightings in state per 10,000 people.

```python
State_Population_Sightings['SightingPerPerson'] = State_Population_Sightings['Posted']/State_Population_Sightings['Population']*10000
State_Population_Sightings
```

The above dataframe is stored into a csv file 'State_Population_Sightings' for later user to plot a heatmap for sightings per 10,000 persons for each state.

```python
State_Population_Sightings.to_csv(r'clean_data/State_Population_Sightings', header=True)
```

# Movie Per Person Data

We begin by reading the total number of scifi movies vs sightings per year from the cleaned scifi_v_sightings.csv. This cleaned dataset contains two columns: 
- 'Year', which is also the index 
- 'Movies' which is the total no of scifi movies
- 'Yearly Sightings' which is the total sightings posted count

```python
yearly_scifiMovies_per_person = pd.read_csv(r'clean_data/scifi_v_sightings', index_col='Year')
```

```python
yearly_scifiMovies_per_person.head()
```

Here, we join the population dataset Pop with yearly_scifiMovies_per_person.

```python
movies_pop_sightings = Pop.join(yearly_scifiMovies_per_person)
```

We drop any NA columns/rows from the above dataset

```python
movies_pop_sightings.head()
movies_pop_sightings = movies_pop_sightings.dropna()
```

```python
movies_pop_sightings.head()
```

The ratio for no of sightings per total population is calcuated for every year.

```python
movies_pop_sightings['SightingPerPerson'] = movies_pop_sightings['Yearly Sightings']/movies_pop_sightings['Population']*10000
```

```python
movies_pop_sightings.head()
```

The ratio for no of movies per total population is calcuated for every year.

```python
movies_pop_sightings['MoviesPerPerson'] = movies_pop_sightings['Movies']/movies_pop_sightings['Population']*10000
```

```python
movies_pop_sightings.head()
```

A new dataframe movies_pop_sightings_ratio is being created to save these ratios - movies per person and sightings per person.

```python
movies_pop_sightings1 = movies_pop_sightings['SightingPerPerson']
```

```python
movies_pop_sightings_ratio = pd.DataFrame(movies_pop_sightings1)
```

```python
movies_pop_sightings_ratio['MoviesPerPerson']=movies_pop_sightings['MoviesPerPerson']
```

```python
movies_pop_sightings_ratio.head()
```

```python
movies_pop_sightings_ratio.to_csv(r'clean_data/movies_pop_sightings_ratio', header=True)
```

The above dataframe is stored into a csv file 'movies_pop_sightings' for later user to plot line graph against years.
