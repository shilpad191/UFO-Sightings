## Difference between Posting and Sighting Date

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from pylab import rcParams
```

```python
plt.style.use('ggplot')
rcParams['figure.figsize'] = 16, 5
```

```python
date_diff = pd.read_csv(r'Clean_data\Post_Sight_Diff')
```

```python
bins = np.arange(0, max(date_diff['Difference in Days']), 10)
plt.hist(date_diff['Difference in Days'], bins);
plt.xlim(0,100);
plt.xlabel("Difference in Post and Sight Date");
plt.ylabel("No of sightings");
```

In the graph above, the difference in the posted and sighting date is incremented by 10 days. The difference is binned in the increments of 10. As you can see in the graph above, the no of sightings is high for the difference of 0 to 10 days. This tells us that people who witniss these sightings will report them online relatively quickly.

## Hourly Sightings without Bins

```python
hourly_sightings = pd.read_csv(r'Clean_data/Hourly_sightings')
```

```python
hourly_sightings.ix[:].plot();
#plt.ylim(0, 6000)
plt.ylabel('No of Sightings');
```

As you can see in the graph above, the number of sightings are less during the day. During day people are at work (working hours- 9:00AM to 6:00PM) and due to sunlight, it becomes harder to notice any unusual lights. Interestingly, there is a gradual increase in the number of sightings during the night-time. Majority of the sightings are in the form of bright light and hence visibility of such unsual light is more during the night time. Not to forget - drinking hours are also during 6:00 PM to 11:00 PM.;-)
